## customer rating in different segments

```python
def customer_rating (Rating):
    if Rating < 4:
        return "Poor"
    elif Rating < 6:
        return "Average"
    elif Rating < 8:
        return "Good"
    elif Rating < 9:
        return "Best"
    else:
        return "Elite"
```

```python
sales_df["customer_rating"]=sales_df["Rating"].apply(customer_rating)
```

```python
sales_df.customer_rating.value_counts()
```

```python
sales_df
```
